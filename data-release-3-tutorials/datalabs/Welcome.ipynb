## Gaia ESA Archive tutorials

When you start a new instance of the Gaia DR3 datalab, the [Github repository](https://github.com/esa/gaia-jupyter-notebooks/tree/main) corresponding to the Data Release 3 Archive tutorials will be cloned into your personal workspace, in the path `/my_worskpace/gaia-jupyter-notebooks` (see below for more details about Workspaces in ESA Datalabs). If the folder already exists, it will be renamed by adding a timestamp to avoid overwritting any local changes.

This repository contains a set of programmatic tutorials in the form of Jupyter notebooks that demonstrate how to access, retrieve, and explore different datasets served by the Gaia ESA Archive. Each notebook in this repository is self-contained and includes detailed instructions and examples.

| Notebook | Description |
|----------|----------|
|[Transform input & output units](../../../my_workspace/gaia-jupyter-notebooks/data-release-3-tutorials/tutorial_change_units.ipynb)| This notebook shows how to use the [Astropy.units](https://docs.astropy.org/en/stable/units/index.html) Python module to transform the units associated to different fields in several Gaia DR3 catalogues. For more information, see also the [Astropy equivalencies](https://docs.astropy.org/en/stable/units/equivalencies.html) documentation.|
|[Retrieve the DataLink products associated to a sample](../../../my_workspace/gaia-jupyter-notebooks/data-release-3-tutorials/tutorial_datalink_download.ipynb)|This notebook shows how to retrieve the different DataLink products from an input list of Gaia DR3 sources. It shows how to access to the data and metadata of the products retrieved in INDIVIDUAL data structure. For simplicity, the files are downloaded in [VOTable](https://www.ivoa.net/documents/VOTable/20191021/) format |
|[Download DataLink products for >5000 sources](../../../my_workspace/gaia-jupyter-notebooks/data-release-3-tutorials/tutorial_datalink_large_samples.ipynb)|This Jupyter Notebook allows to overcome the Gaia Archive DataLink products download threshold (5,000 sources) by first splitting an input source list into multiple chunks of 500 sources each.|
|[Retrieve the BP/RP (XP) spectra associated to a sample of candidate Pleiades stars](../../../my_workspace/gaia-jupyter-notebooks/data-release-3-tutorials/tutorial_use_case_pleiades.ipynb)|This notebook shows how to retrieve the different DataLink products associated to a sample of stars associated to the Pleiades open cluster.|

## The Gaia DR3 Datalab

This Welcome notebook provides helpful information to get started with accessing and analysing data from the **ESA Gaia Mission** in **ESA Datalabs**. 

A **datalab** is a purpose-built software environment that runs on ESA Datalabs. From the user's perspective, it functions similarly to a virtual machine, allowing users to launch, use, and delete it as needed. The **Gaia-DR3 datalab** is based on [JupyterLab](https://jupyterlab.readthedocs.io/en/stable/).

The goal of the **Gaia-DR3** datalab is to help Gaia users become familiar with the ESA Datalabs platform and its core functionalities. In this sense, this initial release is intended as preparation for **Gaia-DR4**, when users will be able to access and analyse data directly within the datalab without needing to download any heavy datasets.

For any questions related to the use of **ESA Datalabs**, please refer to the extensive [Datalabs Help Pages](https://datalabs.esa.int/help/Help.html). 

### Installed packages

This datalab includes a Jupyter kernel (i.e., a conda environment) called **GAIA-DR3**, which has the following pre-installed packages:`Astroquery` `Astropy` `Astropy-healpix` `GaiaXPy` `Healpy` `Kepmodel` `Matplotlib` `Numpy` `Pandas` `Pyesasky` `Pystrometry` `Scipy` `Seaborn`. 

⚠ When you open a Jupyter notebook in the Gaia-DR3 datalab, the GAIA-DR3 kernel should be active by default. **This sould be verified by the user before working with the notebooks**. Otherwise, you can easily activate it using the Switch Kernel option on the right side of the top menu.

To retrieve the full list of installed packages, you can run `%conda list` in a separate cell, provided that the **GAIA-DR3** kernel is activated. 

**Installing new packages**: This datalab has been preparaded to allow users to install additional packages on top of the **GAIA-DR3** pre-built kernel. You can install packages using either the terminal or a Jupyter notebook. However, please note that when you delete the current datalab instance, all installed packages will be removed.

In order to reproduce your custom environment in a future session, the best solution is to create a [conda environment](https://docs.conda.io/projects/conda/en/latest/user-guide/tasks/manage-environments.html) file that can be loaded each time you launch a new datalab instance. For more information about installing packages in Datalabs, see guide [here](https://datalabs.esa.int/help/Getting_started_with_Jupyter_datalabs.html). Environment files can be also shared with collaborators on team workspaces.

### Resources
A user can run **two datalabs** simultaneously. For launching more datalabs, one of the running instances must be deleted. 

When running the Gaia-DR3 datalab, you have access to **32 GB RAM** and **2 CPU cores**. 

**Monitor usage**: For general usage in Datalabs, click **User Icon → My Account → Datalabs Tab**. Additionally, the current total **memory usage** of all kernels and terminals is shown at the bottom left of the Jupyterlab, which is refreshed every 5 seconds. It is also possible to type `diagnose` in a terminal to get a more in-depth analysis.

## Workspaces

ESA Datalabs provides you access to two main workspaces: a private workspace (`/my_workspace`) and a collaborative workspace (/`team_workspaces`):

Both workspaces are available in the user's running datalab and are accesible through the Jupyterlab File Browser. Each user workspace and team workspace is by default allocated **100GB storage**.

⚠ **Do not store files in the datalab itself**, as they will be removed when you delete the running datalab instance. You should copy them to `/my_workspace` or to any of your /`team_workspaces`, which are meant for permanent storage.


## How to access Gaia data in ESA Datalabs

### Gaia data volumes

Gaia DR3 data is available on ESA Datalabs as an external data volume. This volume is automatically mounted when the datalab is started. Direct access to the data is provided in the following paths:

- **Gaia EDR3**: `/data/gaia_datalabs/gedr3` (data volume **1.6 TB**)
- **Gaia DR3**: `/data/gaia_datalabs/gdr3` (data volume **9 TB**)
